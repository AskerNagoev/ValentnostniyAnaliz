```python
grouped = filtered_df1.groupby('data_id')

unique_data = filtered_df1['data_id'].unique()
df_audio_1 = grouped.get_group(unique_data[0])
df_audio_2 = grouped.get_group(unique_data[1])
```

```python
import librosa

audio_path = "в8_гр1_общий.mp3"
audio1, sr = librosa.load(audio_path, sr=22050)
audio_path = "в8_гр2.mp3"
audio2, sr = librosa.load(audio_path, sr=22050)
```

```python
from IPython.display import Audio
Audio(audio2, rate=sr)
```

```python
def slice_audio_to_array(df, audio, sr):
    audio_slices = []
    for idx, row in df.iterrows():
        start_sample = int(row['start'] * sr)  # перевод секунд в семплы
        end_sample = int(row['end'] * sr)      # перевод секунд в семплы
        audio_slice = audio[start_sample:end_sample]
        audio_slices.append(audio_slice)
    return audio_slices
```

```python
audio_slices_1 = slice_audio_to_array(df_audio_1, audio1, sr)

audio_slices_2 = slice_audio_to_array(df_audio_2, audio2, sr)
```

```python
print(f"Total slices in audio_slices_1: {len(audio_slices_1)}")
print(f"Total slices in audio_slices_2: {len(audio_slices_2)}")
```

```python
combined_audio_slices = []

combined_audio_slices.extend(audio_slices_1)
combined_audio_slices.extend(audio_slices_2)

print(f"Total slices in combined_audio_slices: {len(combined_audio_slices)}")
```

```python
spectrogram_list = []

for audio_slice in combined_audio_slices:
    spectrogram = librosa.feature.melspectrogram(y=audio_slice, sr=sr, n_mels = 30)
    spectrogram_flat = spectrogram.flatten()
    spectrogram_list.append(spectrogram_flat)
```

```python
df_spectrogram = pd.DataFrame(spectrogram_list)
```

```python
df_spectrogram.to_csv('hyperc_audio_spectrogram.csv')
```

```python
mfcc_list = []

for audio_slice in combined_audio_slices:
    mfcc = librosa.feature.mfcc(y=audio_slice, sr=sr, n_mfcc = 30)
    mfcc_flat = mfcc.flatten()
    mfcc_list.append(mfcc_flat)

df_mfcc = pd.DataFrame(mfcc_list)
```

```python
df_mfcc.to_csv('hyperc_audio_mfcc.csv')
```

```python
import os
import matplotlib.pyplot as plt
output_image_dir = 'output_images'
os.makedirs(output_image_dir, exist_ok=True)
```

```python
melspectrogram_list = []

for i, audio_slice in enumerate(combined_audio_slices):
    melspectrogram = librosa.feature.melspectrogram(y=audio_slice, sr=sr, n_mels=128, fmax=8000)
    S_dB = librosa.power_to_db(melspectrogram, ref=np.max)
    melspectrogram_list.append(os.path.join(output_image_dir, f'/melspectrogram_{i}.png'))

    # Plotting and saving the melspectrogram
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel', fmax=8000)
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel-frequency spectrogram')
    plt.tight_layout()
    plt.savefig(os.path.join(output_image_dir, f'melspectrogram_{i}.png'))
    plt.close()
```

```python
melspectrogram_path = pd.DataFrame(melspectrogram_list)
melspectrogram_path.to_csv('melspectrogram_path.csv', index=False, header=False)
```
