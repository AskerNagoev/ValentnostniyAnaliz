```python
import nltk
import pymorphy2
from nltk.tokenize import word_tokenize
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer("russian")
nltk.download("stopwords")
stop_words = set(stopwords.words('russian'))
morph = pymorphy2.MorphAnalyzer()
```

```python
def tokenize_text(text):
    tokens = word_tokenize(text)
    return tokens

def lemmatize_tokens(tokens):
    lemmatized_words = [morph.parse(token)[0].normal_form for token in tokens]
    return lemmatized_words

def stem_tokens(tokens):
    stemmed_words = [stemmer.stem(token) for token in tokens]
    return stemmed_words

def remove_stopwords(tokens):
    filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
    return filtered_tokens
```

```python
df_TF_IDF = df
```

```python
df_TF_IDF['text'] = df_TF_IDF['text'].apply(tokenize_text)
df_TF_IDF['text'] = df_TF_IDF['text'].apply(stem_tokens)
df_TF_IDF['text'] = df_TF_IDF['text'].apply(remove_stopwords)
df_TF_IDF = df_TF_IDF[df_TF_IDF['text'].apply(lambda x: len(x) > 0)]
df_TF_IDF
```

```python
word_list = df_TF_IDF['text'].tolist()
text_list = [' '.join(sublist) for sublist in word_list]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=1000)
```

```python
features = vectorizer.fit_transform(text_list)
hyperc_tfidf = pd.DataFrame(features.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
hyperc_tfidf
```

```python
from gensim.models import FastText
import numpy as np
fasttext_model = FastText(sentences=word_list, vector_size=300, window=5, min_count=1, workers=4)
```

```python
fasttext_model.train(word_list, total_examples=len(word_list), epochs=300)
```

```python
fasttext_model.wv.most_similar(positive=["любить"])
```

```python
word_vectors = {}
for word in fasttext_model.wv.key_to_index:
    word_vectors[word] = fasttext_model.wv[word]
```

```python
def average_vector(sublist, word_vectors):
    sum_vector = np.zeros_like(list(word_vectors.values())[0])
    valid_word_count = 0

    for word in sublist:
        if word in word_vectors:
            sum_vector += word_vectors[word]
            valid_word_count += 1

    if valid_word_count == 0:
        return sum_vector

    avg_vector = sum_vector / valid_word_count
    return avg_vector
```

```python
average_vectors = []
for sublist in word_list:
    avg_vec = average_vector(sublist, word_vectors)
    average_vectors.append(avg_vec)
```

```python
df_TF_IDF['fvectors'] = average_vectors
```

```python
hyperc_tfidf.to_csv('hyperc_stem_tfidf.csv', index=False)
df_TF_IDF['valence'].to_csv('hyperc_stem_valence.csv', index=False)

average_vectors_array = np.array(df_TF_IDF['fvectors'].tolist())
np.savetxt("hyperc_stem_fasttext.csv", average_vectors_array, delimiter=",")
```
